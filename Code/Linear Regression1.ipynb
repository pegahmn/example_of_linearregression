```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

```python
x = np.random.rand(100)*20
```

```python
b0 = 7
```

```python
b1 = 16.89
```

```python
noise = np.random.uniform(-5 , 5 , 100)
```

```python
y0 = b0 + b1 * x 
```

```python
y1 = b0 + b1 * x + noise 
```

y with noise

```python
plt.plot(x , y1 ,"o")
plt.show()
```

```python
plt.plot(x , y1 ,"o")
plt.plot(x , y0)
plt.show()
```

```python
plt.plot(x+noise , y0 ,"o")
plt.show()
```

```python
plt.plot(x+noise , y0 ,"o")
plt.plot(x , y0)
plt.show()
```

```python
lm = LinearRegression()
```

```python
lm.fit(x.reshape(-1,1),y1)
```

```python
y_f = lm.predict(x.reshape(-1,1))
```

```python
plt.plot(x,y_f , "blue")
plt.plot(x,y0,"orange")
plt.show
```

```python
lm.score(x.reshape(-1,1),y_f)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
mae = metrics.mean_absolute_error(y_f, y0)
```
